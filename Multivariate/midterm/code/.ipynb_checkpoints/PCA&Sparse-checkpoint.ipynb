```python
import pandas as pd
import numpy as np
from scipy.sparse.linalg import eigs
```

```python
data = pd.read_csv('m-apca40stocks.txt', sep=" ") 
```

```python
data
```

```python
from sklearn.decomposition import PCA           
#加载PCA算法包
pca=PCA(n_components=2,whiten=False)     
#加载PCA算法，设置降维后主成分数目为2
# do not set whiten True because we want the linear combination of returns(features)
reduced_H=pca.fit_transform(data)
# Make sure that pca is of the same length with original data row (# of observations) instead of column (# of stocks)
```

```python
pd.DataFrame(reduced_H).plot()
```

```python
eigenvec = pca.components_[0,:]
```

```python
pd.DataFrame(eigenvec).plot()
```

```python
np.dot(eigenvec, eigenvec)
```

```python
np.argmax(np.abs(eigenvec))
```

```python
print('The explained variance ratios are: {}'.format(pca.explained_variance_ratio_))
```

```python
reduced_H[:,0].var()
```

```python
pd.DataFrame(reduced_H)
```

```python
from sklearn.decomposition import SparsePCA
# 设置成两个pca，得到的component_显示第一个pca主要由第26个股票构成，以下画图对比之
transformer = SparsePCA(n_components=2)
transformer.fit(data)
X_transformed = transformer.transform(data)
```

```python
pd.DataFrame(X_transformed[:,0]).plot()
pd.DataFrame(-data.iloc[:,25]).plot()
```

```python
# most values in the components_ are zero (sparsity)
np.mean(transformer.components_ == 0)
```

```python
np.argmax(np.abs(transformer.components_[0]))
```
